```python
import requests
```

```python
# Simple Json Response
url = 'http://127.0.0.1:8000/predict'
files = {'inputImage': open('../cali_merl_project/For_Testing/41_2_HandinShelf_0_1597_1750.mp4', 'rb')}
response = requests.post(url, files=files)
```

```python
print(response.json())
```

```python
# Return the video with the predicted label on it !
url = 'http://127.0.0.1:8000/predict_w_clip'
video_name = '41_2_HandinShelf_0_1597_1750.mp4'
files = {'inputImage': open(f'../cali_merl_project/For_Testing/{video_name}', 'rb')}
response = requests.post(url, files=files)
```

```python
f = open(f"../api/pred_{video_name}", "wb")
f.write(response.content)
f.close()
```

```python
from moviepy.editor import *
clip = VideoFileClip(f'../api/pred_{video_name}')
    
# Overlay the text clip on the first video clip 
video = CompositeVideoClip([clip])
video.ipython_display(width = 580)
```
